```python
with open("input.txt", "r") as f:
    data = f.readlines()
data
```

```python
my_list = [d.split(": ") for d in data]
my_list
```

```python
#part 1
valid = 0
for requirement, pword in my_list:
    num = requirement.split()[0]
    num1, num2 = num.split("-")
    num1 = int(num1)
    num2 = int(num2)
    
    char = requirement.split()[1]
    
    counter = 0
    for c in pword:
        if c == char:
            counter += 1
    
    if num1 <= counter <= num2:
        valid += 1
            
valid
```
