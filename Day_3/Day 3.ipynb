```python
## Part 2 

#Keep separate position idenitifers for each of the slopes
import time 

start_time = time.time()



c1, c2, c3, c4, c5 = [0]*5

pos_list = [0]*5
pos1, pos2, pos3, pos4, pos5 = pos_list

sign = -1

for line in treemap:
    
    if sign == -1:
        if line[pos5] == "#":
            c5 += 1
        pos5 += 1
    
    sign *= -1
    
    if line[pos1] == "#":
        c1 += 1
    
    if line[pos2] == "#":
        c2 += 1
    
    if line[pos3] == "#":
        c3 += 1
    
    if line[pos4] == "#":
        c4 += 1
        
    pos1 += 1
    pos2 += 3
    pos3 += 5
    pos4 += 7
    
    if pos1 > 30:
        pos1 -= 31
    
    if pos2 > 30:
        pos2 -= 31
    
    if pos3 > 30:
        pos3 -= 31
    
    if pos4 > 30:
        pos4 -= 31
        
    if pos5 > 30:
        pos5 -= 31
            
c1 * c2 * c3 * c4 * c5

time_passed = time.time() - start_time
time_passed
```
